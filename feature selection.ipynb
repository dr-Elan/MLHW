### 1. Сгенерируйте данные с помощью кода

```python
from sklearn.datasets import make_classification
x_data_generated, y_data_generated = make_classification(scale=1)
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
from sklearn.model_selection import cross_val_score
```

### 2. Постройте модель логистической регрессии и оцените среднюю точность.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_data_generated, y_data_generated)
cross_val_score(LogisticRegression(), x_data_generated, y_data_generated, scoring='accuracy').mean()
```

### 3. Используйте статистические методы для отбора признаков

```python
df = pd.DataFrame(x_data_generated)
df.corr()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,15))         # Sample figsize in inches
sns.heatmap(df.corr(), annot=True, linewidths=.5, ax=ax)
```

```python
from sklearn.feature_selection import VarianceThreshold
import matplotlib.pyplot as plt
```

```python
threshol = VarianceThreshold(threshold=1)
x_high_variance = threshol.fit_transform(x_data_generated)

df_thre = pd.DataFrame(x_high_variance)
fig, ax = plt.subplots(figsize=(15,15))         # Sample figsize in inches
sns.heatmap(df_thre.corr(), annot=True, linewidths=.5, ax=ax)
```

```python
model = LogisticRegression()
model.fit(x_high_variance, y_data_generated)
cross_val_score(LogisticRegression(), x_high_variance, y_data_generated, scoring='accuracy').mean()
```

### 4. Осуществите отбор признаков на основе дисперсионного анализа

```python
from sklearn.feature_selection import SelectKBest, chi2, f_classif
```

```python
X_new_shape = SelectKBest(f_classif, k = 5 ).fit_transform(x_data_generated, y_data_generated)
```

```python
model = LogisticRegression()
model.fit(X_new_shape, y_data_generated)
cross_val_score(LogisticRegression(), X_new_shape, y_data_generated, scoring='accuracy').mean()
```
