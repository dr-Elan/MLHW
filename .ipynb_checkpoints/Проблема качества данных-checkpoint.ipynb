```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import roc_curve
le = LabelEncoder()
ohe = OneHotEncoder()
import scipy.stats as stats
import re
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model = LogisticRegression()
```

### 2. Удалите все пропущенные и категориальные переменные. Обучите модель. Выберете и посчитайте метрику качества

```python
df.info()
```

```python
df.nunique()
```

```python
# удалим категориальные переменные
df = df[['Survived','Sex','Age','Fare']]
```

```python
# отобразим пропуски
df.isnull().sum()
sns.heatmap(df.isnull(), cbar = False).set_title("Missing values heatmap")
```

```python
df = df.dropna()
df
```

```python
X = pd.get_dummies(df[['Sex','Age','Fare']])
X
```

```python
le.fit(df['Survived'])
```

```python
y = pd.Series( data = le.transform( df['Survived'] ) )
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model.fit(X_train, y_train)
```

$$
AccuracyScore(y,\hat{y}) = \frac{1}{n_{sample}} \sum\limits_0^{n_{sample}} 1(y=\hat{y})
$$

```python
from sklearn.metrics import accuracy_score
accuracy_score(model.predict(X_train),y_train)
```

### 7. Проверьте данные на наличие выбросов.
#### a) Удалите выбросы, если считаете это целесообразным. Обоснуйте.

Z-оценка показывает, сколько стандартных отклонений данного значения от среднего. Мы используем следующую формулу для расчета z-показателя:

$$
Z = \frac{X - \mu}{\sigma}
$$

```python
z = np.abs(stats.zscore(df[['Age','Fare']]))
z
```

```python
df[['Age','Fare']].describe()
```

```python
df[['Age','Fare']][(z<5).all(axis=1)].describe()
```
